```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(10);
```

```python
mean = (0, 0)
cov1 = [[1, 0],
       [0, 1]]
cov2 = [[1, 0],
       [0, 2]]
cov3 = [[1, -1],
       [-1, 1]]
cov4 = [[1, 0.5],
       [0.5, 2]]
covs = [cov1, cov2, cov3, cov4]
fig, axs = plt.subplots(2,2,figsize=(7,7), sharex=True, sharey=True)
k=0
for idx, cov in enumerate(covs):
    data = np.random.multivariate_normal(mean, cov, 500)
    # plt.scatter(data[:,0], data[:,1])
    r = k/2
    c = k%2
    plt.sca(axs[r,c])
    sns.kdeplot(data[:,0], data[:,1], bw=.99)#, cmap="Reds", shade=False, bw=.15)
    plt.xlabel(r'$x_1$')
    plt.ylabel(r'$x_2$')
    plt.title('ABCD'[k], fontsize=20)
    plt.axhline(0, color='black')
    plt.axvline(0, color='black')
    plt.xlim([-max(np.abs(plt.axis())), max(np.abs(plt.axis()))])
    plt.ylim([-max(np.abs(plt.axis())), max(np.abs(plt.axis()))])
    plt.grid()
    k+=1
```
